# Análise de Homologias por Blast da proteína Wolframin

```python
from Bio.Blast import NCBIXML 
from Bio.Blast import NCBIWWW 
from Bio import SeqIO
```

Importar a sequência relativa à Wolframin

```python
record = SeqIO.read(open("prot_wsf1.fasta"),"fasta") 
seq = record.seq
print (len(seq)) 


# 890 aa  
```

Realizar o protein blast da proteína com a base de dados nr e guardar num ficheiro xml

```python
result_handle = NCBIWWW.qblast("blastp", "nr", record.format("fasta"))
```

```python
file_save = open("blastpwsf1.xml", "w")
file_save.write(result_handle.read()) 
file_save.close() 
result_handle.close()
```

Abrir e ler o resultado guardado no ficheiro xml

```python
result_handle = open("blastpwsf1.xml")
record =  NCBIXML.read(result_handle)
```

## Parsing dos resultados

Visualizar os parâmetros globais da pesquisa

```python
print("Parâmetros")
print("Database: " , record.database)
print("Matrix: ", record.matrix)
print("Gap penalties: ", record.gap_penalties)
```

Visualizar as sequências homólogas encontradas, as suas características e o alinhamento, sendo o critério todas aquelas que apresentarem valor de E value superior a 0.00001, query cover superior a 95% e percent idenitity superior a 98,5% 

```python
E_VALUE_THRESHOLD=0.00001
query_threshold=0.95
identity_threshold=0.985
tam=len(seq) #tamanho
ct = 0
print('******Sequências Homólogas******')
for alignment in record.alignments:
    for hsp in alignment.hsps:
        ct+=1
        tam_ali=hsp.align_length
        query=(tam_ali/tam)
        identity=(hsp.identities)/tam_ali
        if hsp.expect < E_VALUE_THRESHOLD and query > query_threshold and identity > identity_threshold: 
            print('\n','****Sequência',ct,'****')
            print('Accession:' ,alignment.accession)
            print('Description:',alignment.hit_def)
            print('Description: wolframin [Homo sapiens]')
            print('Length:',alignment.length)
            print('E-value:', hsp.expect)
            print('Max Score: ', hsp.bits)
            print('Query cover: ', (query*100) ,'%')
            print('Percent identity: ', ((identity)*100),'%' )
            print('Positivies:', (((hsp.positives)/tam_ali)*100), '%' ,'\n')
            print('**Alignment**')
            print(hsp.query[0:125])
            print(hsp.match[0:125])
            print(hsp.sbjct[0:125],'\n')
            print(hsp.query[125:250])
            print(hsp.match[125:250])
            print(hsp.sbjct[125:250],'\n')
            print(hsp.query[250:375])
            print(hsp.match[250:375])
            print(hsp.sbjct[250:375],'\n')
            print(hsp.query[375:500])
            print(hsp.match[375:500])
            print(hsp.sbjct[375:500],'\n')
            print(hsp.query[500:625])
            print(hsp.match[500:625])
            print(hsp.sbjct[500:625],'\n')
            print(hsp.query[625:750])
            print(hsp.match[625:750])
            print(hsp.sbjct[625:750],'\n')
            print(hsp.query[750:875])
            print(hsp.match[750:875])
            print(hsp.sbjct[750:875],'\n')
            print(hsp.query[875:890])
            print(hsp.match[875:890])
            print(hsp.sbjct[875:890],'\n')
            
            
```
